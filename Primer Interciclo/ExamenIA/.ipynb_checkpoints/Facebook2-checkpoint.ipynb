```python
from neomodel import StructuredNode, StringProperty, RelationshipTo, RelationshipFrom, config, IntegerProperty

config.DATABASE_URL = 'bolt://neo4j:1234@localhost:7687'

class Partido(StructuredNode):
    nombre = StringProperty(unique_index=True)
    
class Presidente(StructuredNode):
    nombre = StringProperty(unique_index=True)
    presidente = RelationshipTo('Partido', 'candidato presidencial')
    
class Vicepresidente(StructuredNode):
    nombre = StringProperty(unique_index=True)
    vicepresidente = RelationshipTo('Partido', 'candidato vicepresidencial')

class Asambleista(StructuredNode):
    nombre = StringProperty(unique_index=True)
    asambleista = RelationshipTo('Partido', 'asambleista')
    
class Parurl(StructuredNode):
    url = StringProperty(unique_index=True)
    publicacionUrl = RelationshipTo('Partido', 'pertenece')
    
class Parlink(StructuredNode):
    text = StringProperty(unique_index=True)
    likes = IntegerProperty(index=True)
    time = StringProperty(index=True)
    publicacionInf = RelationshipTo('Parurl', 'informacion')    
    
class Preurl(StructuredNode):
    url = StringProperty(unique_index=True)
    presidenteUrl = RelationshipTo('Presidente', 'pertenece')
    
class Prelink(StructuredNode):
    text = StringProperty(unique_index=True)
    likes = IntegerProperty(index=True)
    time = StringProperty(index=True)
    presidenteInf = RelationshipTo('Preurl', 'informacion')  
    
class Viceurl(StructuredNode):
    url = StringProperty(unique_index=True)
    viceUrl = RelationshipTo('Vicepresidente', 'pertenece')
    
class Vicelink(StructuredNode):
    text = StringProperty(unique_index=True)
    likes = IntegerProperty(index=True)
    time = StringProperty(index=True)
    viceInf = RelationshipTo('Viceurl', 'informacion')
    
class Asaurl(StructuredNode):
    url = StringProperty(unique_index=True)
    AsaUrl = RelationshipTo('Asambleista', 'pertenece')
    
class Asalink(StructuredNode):
    text = StringProperty(unique_index=True)
    likes = IntegerProperty(index=True)
    time = StringProperty(index=True)
    AsaInf = RelationshipTo('Asaurl', 'informacion')
    
    
```

```python
partido = Partido(nombre='UNES',).save()
presidente = Presidente(nombre='ANDR√âS DAVID ARAUZ GALARZA',).save()
vicepresidente = Vicepresidente(nombre='CARLOS XAVIER	RABASCALL SALAZAR',).save()
asambleista = Asambleista(nombre='JUAN CRIST√ìBAL LLORET VALDIVIESO',).save()
presidente.presidente.connect(partido)
vicepresidente.vicepresidente.connect(partido)
asambleista.asambleista.connect(partido)
```

```python
from facebook_scraper import get_posts

for post in get_posts('UnionPorLaEsperanza', pages=50):
    print(post)
    if (post['post_url'][:1000]) == "":
        print('No se encuentra')
    else:    
        Publicacion = Parurl(url=(post['post_url'][:1000])).save()
        Publicacion.publicacionUrl.connect(partido)
        textoUnes = Parlink(text=(post['text'][:1000]), likes=(post['likes']), time=(post['time'])).save()
        textoUnes.publicacionInf.connect(Publicacion)
        #print("========================================================")
```
